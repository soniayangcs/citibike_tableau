```python
# dependencies
import os
import csv
import pandas as pd
```

```python
# manually using the same code again for each year as opposed to a gigantic loop for all of them
# this is not necessarily ideal, but will allow for testing in small increments
# also less wait time to get results per year
```

<h2>2013</h2>

```python
# list of csv numbers by month
months = ['06', '07', '08', '09', '10', '11', '12']

# empty list to hold dataframes
dfs = []

for month in months:
    citibikeCSV = os.path.join('2013', '2013-' + month +'-CitiBike_tripdata' + '.csv')
    print(citibikeCSV)
    
    df = pd.read_csv(citibikeCSV)
    dfs.append(df)

df =  pd.concat(dfs)
df
```

```python
# export final result to a CSV
df.to_csv('merged/citibike_2013.csv', encoding='utf-8', index=False)
```
